```python
# Define que a autenticação do Azure Data Lake Gen2
# será feita usando OAuth (e não chave de acesso)
spark.conf.set(
  "fs.azure.account.auth.type.armazenamentodatalake26.dfs.core.windows.net",
  "OAuth"
)

# Define o provedor OAuth baseado em Client Credentials
# (Service Principal: Client ID + Client Secret)
spark.conf.set(
  "fs.azure.account.oauth.provider.type.armazenamentodatalake26.dfs.core.windows.net",
  "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider"
)

# Client ID (Application / Service Principal)
# Identifica quem está tentando acessar o Data Lake
spark.conf.set(
  "fs.azure.account.oauth2.client.id.armazenamentodatalake26.dfs.core.windows.net",
   dbutils.secrets.get(scope="adls-scope", key="client-id")
)

# Client Secret do Service Principal
# Funciona como a senha da aplicação
spark.conf.set(
  "fs.azure.account.oauth2.client.secret.armazenamentodatalake26.dfs.core.windows.net",
 dbutils.secrets.get(scope="adls-scope", key="client-secret")
)

# Endpoint de autenticação do Microsoft Entra ID (Azure AD)
# Usa o Tenant ID para gerar o token OAuth
spark.conf.set(
  "fs.azure.account.oauth2.client.endpoint.armazenamentodatalake26.dfs.core.windows.net",
  "https://login.microsoftonline.com/68b68edc-0e87-487d-8b75-8531dc592bf6/oauth2/v2.0/token"
)
```
